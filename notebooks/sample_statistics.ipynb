# Functions

```python
def find_avgs(prop, df): 
    avgs_ls = []
    for feat in feats:
        feat_df = df[df['feat_name'] == feat]
        avg = np.mean(feat_df[prop])
        avgs_ls.append(avg)
    return avgs_ls

def find_stdev(prop, df):
    stds_ls = []
    for feat in feats:
        feat_df = df[df['feat_name'] == feat]
        std = np.std(feat_df[prop])
        stds_ls.append(std)
    return stds_ls
```

```python
def avg_table(prop_list, df):
    avg_df = pd.DataFrame()
    avg_df['feature'] = feats
    for prop in prop_list:
        avg_column = find_avgs(prop, df)
        std_column = find_stdev(prop, df)
        avg_df[f'mean_{prop}'] = avg_column
        avg_df[f'stdev_{prop}'] = std_column
        
    return avg_df
```

```python
subtypes = ['SS', 'BL', 'CN', 'CL']
prop_list2 = ['zCMB','logMassFSPS','logSSFRFSPS','ageFSPS','kpc', 'x0SALT2zspec', 'x1SALT2zspec', 'cSALT2zspec'
              , 'PeakMJDSALT2zspec', 'muSALT2zspec']
```

```python
def find_avgs2(prop, df): 
    avgs_ls = []
    
    total_avg = np.mean(df[prop])
    avgs_ls.append(total_avg)
    
    for subtype in subtypes:
        sub_df = df[df['branch_type'] == subtype]
        avg = np.mean(sub_df[prop])
        avgs_ls.append(avg)
        
    return avgs_ls
```

```python
def find_stdev2(prop, df):
    stds_ls = []
    
    total_std = np.std(df[prop])
    stds_ls.append(total_std)
    
    for subtype in subtypes:
        sub_df = df[df['branch_type'] == subtype]
        std = np.std(sub_df[prop])
        stds_ls.append(std)
        
    return stds_ls
```

```python
def avg_table2(prop_list, df):
    
    avg_df = pd.DataFrame()
    data_types = ['All','SS', 'BL', 'CN', 'CL']
    avg_df['data_type'] = data_types
    
    for prop in prop_list:
        avg_column = find_avgs2(prop, df)
        std_column = find_stdev2(prop, df)
        avg_df[f'mean_{prop}'] = avg_column
        avg_df[f'stdev_{prop}'] = std_column
        
    return avg_df
```
