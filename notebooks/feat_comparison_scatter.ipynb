```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MultipleLocator
from scipy.stats.stats import pearsonr
from pandas import DataFrame
```

```python
path = '/Users/emilysf098/Desktop/Research/SN-Spectral-Evolution'
sdss_combined = pd.read_csv(path + '/results/sdss_combined.csv')
```

```python
CIDs = list(set(sdss_combined['obj_id']))
feats = ['pW1', 'pW2', 'pW3', 'pW4', 'pW5', 'pW6', 'pW7']
```

```python
def two_feat_table(feat1, feat2):
    valid_obj = []
    for cid in CIDs:
        df = sdss_combined[sdss_combined['obj_id'] == cid]
        if feat1 in list(df['feat_name']) and feat2 in list(df['feat_name']):
            valid_obj.append(cid)
        else:
            pass
    indexed_df = sdss_combined.set_index('obj_id')
    new_df = indexed_df.loc[valid_obj, :]
    return new_df[new_df['feat_name'] == feat1] , new_df[new_df['feat_name'] == feat2]
```

```python
one_one_ones, one_one_ones = two_feat_table('pW1', 'pW1')
one_two_ones, one_two_twos = two_feat_table('pW1', 'pW2')
one_three_ones, one_three_threes = two_feat_table('pW1', 'pW3')
one_four_ones, one_four_fours = two_feat_table('pW1', 'pW4')
one_five_ones, one_five_fives = two_feat_table('pW1', 'pW5')
one_six_ones, one_six_sixs = two_feat_table('pW1', 'pW6')
one_sev_ones, one_sev_sevs = two_feat_table('pW1', 'pW7')

two_two_twos, two_two_twos = two_feat_table('pW2', 'pW2')
two_three_twos, two_three_threes = two_feat_table('pW2', 'pW3')
two_four_twos, two_four_fours = two_feat_table('pW2', 'pW4')
two_five_twos, two_five_fives = two_feat_table('pW2', 'pW5')
two_six_twos, two_six_sixs = two_feat_table('pW2', 'pW6')
two_sev_twos, two_sev_sevs = two_feat_table('pW2', 'pW7')

three_three_threes, three_three_threes = two_feat_table('pW3', 'pW3')
three_four_threes, three_four_fours = two_feat_table('pW3', 'pW4')
three_five_threes, three_five_fives = two_feat_table('pW3', 'pW5')
three_six_threes, three_six_sixs = two_feat_table('pW3', 'pW6')
three_sev_threes, three_sev_sevs = two_feat_table('pW3', 'pW7')

four_four_fours, four_four_fours = two_feat_table('pW4', 'pW4')
four_five_fours, four_five_fives = two_feat_table('pW4', 'pW5')
four_six_fours, four_six_sixs = two_feat_table('pW4', 'pW6')
four_sev_fours, four_sev_sevs = two_feat_table('pW4', 'pW7')

five_five_fives, five_five_fives = two_feat_table('pW5', 'pW5')
five_six_fives, five_six_sixs = two_feat_table('pW5', 'pW6')
five_sev_fives, five_sev_sevs = two_feat_table('pW5', 'pW7')

six_six_sixs, six_six_sixs = two_feat_table('pW6', 'pW6')
six_sev_sixs, six_sev_sevs = two_feat_table('pW6', 'pW7')

sev_sev_sevs, sev_sev_sevs = two_feat_table('pW7', 'pW7')
```
