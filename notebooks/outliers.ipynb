# Investigating Outliers

this notebook plots the spectra of outliers from the branch classification and host gal prop plots

```python
from sndata.sdss import Sako18Spec
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dao = Sako18Spec()
out_1 = dao.get_data_for_id('10449')  
```

```python
#outliers from branch plot
branch_outliers = ['10449', '16021', '1371', '6773', '7147', '16099', 
                   '16281', '18375', '18697', '21502']
redshift_outliers = ['16280', '3901', '17825', '2330', '1580', '15369', '13072', '5737', '5994', '16287',
                     '6422', '6933', '16333']
hostmass_outliers = ['16280', '3901', '2330', '1580', '15369', '6422', '13072', '16333']
sfr_outliers = ['1580', '1253', '20350', '16280', '16333']
```

```python
def show_spectra(spectrum):
    """ shows the spectrum of an object
    Args:
        spectrum     (string): object ID of outlier 
    """
    
    data = dao.get_data_for_id(spectrum)
    wave = data['wavelength']
    flux = data['flux']
    name = spectrum
    
    plt.figure(figsize=(15,7))
    plt.plot(wave, flux, '-ok', linewidth = 1, markersize=0.01, color='black')
    plt.title(spectrum)
```

```python
def show_spectra_close(spectrum, ymin, ymax):
    """incomplete function to zoom in on a spectrum
    Args:
        spectrum (string): object ID of outlier 
        ymin (int): min y value
        ymax (int): max y value
    """
    data = dao.get_data_for_id(spectrum)
    wave = data['wavelength']
    flux = data['flux']
    name = spectrum
    
    plt.figure(figsize=(15,7))
    plt.plot(wave, flux, '-ok', linewidth = 1, markersize=0.01, color='black')
    plt.ylim(ymin, ymax)
    plt.title(spectrum)
```

```python
for outlier in branch_outliers:
    show_spectra(outlier)
```

```python
for outlier in redshift_outliers:
    show_spectra(outlier)
```

```python
for outlier in hostmass_outliers:
    show_spectra(outlier)
```

```python
for outlier in sfr_outliers:
    show_spectra(outlier)
```
