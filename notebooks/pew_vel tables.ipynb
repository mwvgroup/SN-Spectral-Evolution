```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
path = '/Users/emilysf098/Desktop/Research/SN-Spectral-Evolution'
sdss_combined = pd.read_csv(path + '/results/sdss_combined.csv', index_col=['feat_name', 'obj_id'])
sdss_combined = sdss_combined.iloc[: , 1:]
sdss_combined = sdss_combined.reset_index()
#sdss_combined.duplicated(subset=None, keep='first')
```

```python
CID = list(set(sdss_combined['obj_id']))
CID.sort()
pW_list = ['pW1', 'pW2', 'pW3', 'pW4', 'pW5', 'pW6', 'pW7', 'pW8']
```

```python
def make_table_pew():
    pew_ls_ls = []
    pew_err_ls_ls = []
    for cid in CID: #for each object...
        pW_pews = [] #list of feature PEWs in order
        pW_pews_err = [] #list of PEW errors in order
        cid_table = sdss_combined[sdss_combined['obj_id'] == cid] 
        
        for pw in pW_list:
            if pw in cid_table['feat_name'].to_list():
                pew = cid_table['pew'][cid_table['feat_name'] == pw]
                pew_err = cid_table['pew_samperr'][cid_table['feat_name'] == pw]
                pew = pew.values
                pew_err = pew_err.values
                pew = pew.tolist()
                pew_err = pew_err.tolist()
                pew = str(pew).lstrip('[').rstrip(']')
                pew = float(pew)
                pew_err = str(pew_err).lstrip('[').rstrip(']')
                pew_err = float(pew_err)
                pW_pews.append(pew)
                pW_pews_err.append(pew_err)
            else:
                pW_pews.append(np.nan)
                pW_pews_err.append(np.nan)
        
        pew_ls_ls.append(pW_pews)
        pew_err_ls_ls.append(pW_pews_err)
    
        
    return pew_ls_ls, pew_err_ls_ls
   
```

```python
pews_ls, pew_errs_ls = make_table_pew() 
#a list of lists of feature PEWs. EX: first element is the PEWs for each feature for the first object in CID
```

```python
pw1_col = []
pw1_err_col = []
pw2_col = []
pw2_err_col = []
pw3_col = []
pw3_err_col = []
pw4_col = []
pw4_err_col = []
pw5_col = []
pw5_err_col = []
pw6_col = []
pw6_err_col = []
pw7_col = []
pw7_err_col = []
pw8_col = []
pw8_err_col = []

for i in range(len(CID)):
    pews = pews_ls[i]
    if pews != np.nan:
        pw1_col.append(pews[0])
        pw2_col.append(pews[1])
        pw3_col.append(pews[2])
        pw4_col.append(pews[3])
        pw5_col.append(pews[4])
        pw6_col.append(pews[5])
        pw7_col.append(pews[6])
        pw8_col.append(pews[7])
    else:
        pw1_col.append(np.nan)
        pw2_col.append(np.nan)
        pw3_col.append(np.nan)
        pw4_col.append(np.nan)
        pw5_col.append(np.nan)
        pw6_col.append(np.nan)
        pw7_col.append(np.nan)
        pw8_col.append(np.nan)
        
    pews_errs = pew_errs_ls[i]
    if pews_errs != np.nan:
        pw1_err_col.append(pews_errs[0])
        pw2_err_col.append(pews_errs[1])
        pw3_err_col.append(pews_errs[2])
        pw4_err_col.append(pews_errs[3])
        pw5_err_col.append(pews_errs[4])
        pw6_err_col.append(pews_errs[5])
        pw7_err_col.append(pews_errs[6])
        pw8_err_col.append(pews_errs[7])
    else:
        pw1_err_col.append(np.nan)
        pw2_err_col.append(np.nan)
        pw3_err_col.append(np.nan)
        pw4_err_col.append(np.nan)
        pw5_err_col.append(np.nan)
        pw6_err_col.append(np.nan)
        pw7_err_col.append(np.nan)
        pw8_err_col.append(np.nan)

pEW_df = pd.DataFrame(list(zip(CID, pw1_col, pw1_err_col, pw2_col, pw2_err_col, pw3_col, pw3_err_col, pw4_col, pw4_err_col, pw5_col, pw5_err_col, pw6_col, pw6_err_col, pw7_col, pw7_err_col, pw8_col, pw8_err_col)),
                index = None, columns =['CID', 'pW1 pEW', 'pW1 pEW err', 'pW2 pEW', 'pW2 pEW err', 'pW3 pEW',
                                        'pW3 pEW err', 'pW4 pEW', 'pW4 pEW err', 'pW5 pEW', 'pW5 pEW err',
                                        'pW6 pEW', 'pW6 pEW err', 'pW7 pEW', 'pW7 pEW err', 'pW8 pEW',
                                        'pW8 pEW err'], dtype = None)

pEW_df = pEW_df.drop(['pW8 pEW', 'pW8 pEW err'], axis=1)
pEW_df = pEW_df.round(decimals = 3)
pEW_df.to_csv(path + '/results/paper_pEW_df.csv')
```

```python
def make_table_vel():
    vel_ls_ls = []
    vel_err_ls_ls = []
    for cid in CID: #for each object...
        pW_vels = [] #list of feature vels in order
        pW_vels_err = [] #list of vel errors in order
        cid_table = sdss_combined[sdss_combined['obj_id'] == cid] 
        
        for pw in pW_list:
            if pw in cid_table['feat_name'].to_list():
                vel = cid_table['vel'][cid_table['feat_name'] == pw]
                vel_err = cid_table['vel_samperr'][cid_table['feat_name'] == pw]
                vel = vel.values
                vel_err = vel_err.values
                vel = vel.tolist()
                vel_err = vel_err.tolist()
                vel = str(vel).lstrip('[').rstrip(']')
                vel = float(vel)
                vel_err = str(vel_err).lstrip('[').rstrip(']')
                vel_err = float(vel_err)
                pW_vels.append(vel)
                pW_vels_err.append(vel_err)
            else:
                pW_vels.append(np.nan)
                pW_vels_err.append(np.nan)
        
        vel_ls_ls.append(pW_vels)
        vel_err_ls_ls.append(pW_vels_err)
    
        
    return vel_ls_ls, vel_err_ls_ls
   
```
