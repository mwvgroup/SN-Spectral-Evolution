```python
import numpy as np

def get_CSP_t0(obj_id):
    table_3 = dr3.load_table(3)
    if not obj_id in table_3['SN']:
        raise ValueError(f' t0 not available for [obj-id]')
    obj_data = table_3[table_3['SN'] == obj_id]
    return obj_data['T(Bmax)']
    
```

```python
demo_id = object_ids[0]
dr1.get_data_for_id(demo_id)

def get_spectrum_near_t0(obj_id):
    data = dr1.get_data_for_id(obj_id)
    date = np.unique(data['time'])
    t0 = get_CSP_t0(obj_id)
    phase = date - t0
    index = np.argmin(np.abs(phase))
    peak_date = date[index]
    peak_data = data[data['time'] == peak_date]
    return peak_data

plt.figure(figsize = (11, 7))
for i, obj_id in enumerate(dr1.get_available_ids()[0:8]):
    peak_data = get_spectrum_near_t0(obj_id)
    plt.plot(peak_data['wavelength'], peak_data['flux'] + i*1e-16, label = obj_id)
    plt.xlabel("Wavelength [$\AA$]")
    plt.ylabel("flux + C")
    plt.legend(title = "Objects", loc = "upper right")
    plt.title("Observed Spectra at Peak Brightness from CSP DR1")
    
```
