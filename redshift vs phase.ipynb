```python
def intersection(lst1, lst2): 
    '''
    return intersection of two lists
    
    args
    lst1 - first list to be compared
    lst2 - second list to be compared
    
    returns
    (set(lst1) & set(lst2)) - intersection of both lists'''
    return (set(lst1) & set(lst2))
```
